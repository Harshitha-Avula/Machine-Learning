```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.random.random(50)
x
```

```python
y=np.log(x)+3*x
```

```python
plt.plot(x,y,'o')
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x.reshape(-1,1),y)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
plt.plot(x,y,'ro')
plt.plot(x,LR.predict(x.reshape(-1,1)),'bo')
```

```python
df=pd.DataFrame()
df['x1']=x
df['x2']=np.log(x)
df
```

```python
LR1=LinearRegression()
LR1.fit(df,y)
```

```python
# y = 1*x2 + 3*x1 + 0
print(LR1.intercept_,LR1.coef_)
```

```python
plt.scatter(np.log(x),y)
```

```python
plt.scatter(x**3,y)
```

```python
# polynomial feature
from sklearn.preprocessing import PolynomialFeatures
PF = PolynomialFeatures(degree=2)
PF.fit(df[['x1']])
PF.transform(df[['x1']])
```

```python
from sklearn.preprocessing import PolynomialFeatures
PF = PolynomialFeatures(degree=3)
PF.fit(df[['x1']])
PF.transform(df[['x1']])
poly_df = pd.DataFrame(PF.transform(df[['x1']]),columns=['x1','x2','x3','x4'])
```

```python
poly_df
```

```python
poly_df['y'] = y
```

```python
poly_df.corr()
```
