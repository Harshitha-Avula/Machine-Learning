```python
x=np.array([40,30,20,10,100])
q1=np.quantile(x,0.25)
q2=np.quantile(x,0.5)
q3=np.quantile(x,0.75)
```

```python
q1,q2,q3
```

```python
#iqr - inter quantile range
# iqr = q3-q1
#iqr *1.5
#outliers:
#value < q1-iqr*1.5 
#q3+iqr*1.5 > value
```

```python
IQR=q3-q1
(q1-IQR*1.5,q3+IQR*1.5)
```
